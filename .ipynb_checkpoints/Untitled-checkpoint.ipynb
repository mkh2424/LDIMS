```python
import numpy as np
import datetime

def filter_single_crosschecked(df_query):
    # 한사람, 교차검증된 프로파일(조건 : GF/PPF로 두번 확인하고 한 사람의 프로파일일 경우만) 필터링.
    score_query = df_query[standard_locus]
    score_query = score_query.applymap(lambda x: len(str(x).split('-')) if str(x)!='' else 0)
    score_query = score_query.applymap(lambda x: x if x in [1,2] else np.nan)
    idx_passed = score_query.dropna().index
    df_passed = df_query.loc[idx_passed]
    return df_passed

# def filter_overlap(df_query, df_target):
#     # 중복되는 자료 처리
#     SN_query = df_query['Sample Name'].values
#     idx_query_overlapped = []
#     idx_target_overlapped = []
#     for idx_query, SN in enumerate(SN_query):
#         overlapped_target = df_target[df_target['Sample Name']==SN]
#         overlapped_query = SN_query[SN_query==SN]
#         if len(overlapped_target) !=0:
#             print(f'{SN} - already existed Sample Name : {len(overlapped_target)}')
#             idx_target_overlapped.extend(overlapped_target.index) # 항상 갱신.
#     df_target.drop(labels=idx_target_overlapped, axis=0, inplace=True) # 샘플 이름이 중복된 경우 기존 DB 것을 삭제.
#     return df_target
#     # query 추가    
#     #         나중에 하나씩 입력할 때
#     #         dialog_update = input("중복 발견. y: 갱신, n: 넘어가기, p: 중단 - ")
#     #         if dialog_update == 'y':
#     #             idx_target_overlapped.extend(overlapped_target.index)
#     #         elif dialog_update == 'n':
#     #             idx_query_overlapped.append(index_query)
#     #         else:
#     #             break # return

def filter_overlap(df_query, df_target):
    df_sum = pd.concat([df_target, df_query])
    duplicated = df_sum.duplicated(['Sample Name'])
    if sum(duplicated)>0:
        list_duplicated = df_sum[duplicated]['Sample Name'].unique()
        print(f'중복되는 Sample Name이 발견되었습니다. 기존의 데이터를 제거합니다. : {list_duplicated}')    
        df_sum = pd.concat([df_target, df_query, df_query])
        df_target_filtered = df_sum.drop_duplicates(['Sample Name'], keep=False,ignore_index=True)
    else:
        df_target_filtered = df_target
    return df_target_filtered
    
def filter_date(df_target, date_from, date_to):
    # 지정한 기간에 저장된 DB 자료를 필터링
    #date_query1 = datetime.datetime.strptime('20210621', '%Y%m%d')
    df_date_filtered = df_target.copy()
    df_date_filtered['datetime'] = df_date_filtered['Date'].apply(lambda x:datetime.datetime.strptime(x,'%Y%m%d'))
    condition1_date = df_date_filtered['datetime'] > date_from
    condition2_date = df_date_filtered['datetime'] < date_to
    df_date_filtered = df_date_filtered[condition1_date]
    df_date_filtered = df_date_filtered[condition2_date]
    df_date_filtered.drop(['datetime'], axis=1, inplace=True)
    return df_date_filtered

def add_query(df_query, df_target):
    df_query_filterd = filter_single_crosschecked(df_query)
    df_target_filterd = filter_overlap(df_query_filterd, df_target)
    df_sum = pd.concat([df_query_filterd, df_target_filterd], axis=0)
    df_sum.reset_index(inplace=True, drop=True)
    return df_sum
    
```

```python
import Modules.NFS_DNA as NFS_DNA
import pandas as pd
import os

df_DB = pd.read_csv('local_DB.csv', header=0, index_col=0)

path_query = os.getcwd()+'/tomato_part/'
filenames = os.listdir(path_query)
cnt_stored = 0
for filename in filenames:
    try:
        query = NFS_DNA.CombinedResult()
        query.load_tomato(path_query+filename)
    except:
        print(f'Failed to load the tomato file - Filename : {filename}')
    else:
        cnt_stored+=1
        df_DB = add_query(query.df_profiles, df_DB)       
        print(f'{path_query+filename} stored')

df_DB.sort_values(by=['Date'], axis=0, ascending=True, inplace=True)
df_DB.reset_index(drop=True, inplace=True)
df_DB.to_csv("local_DB.csv",sep=',')

print(f'{cnt_stored} files stored')
```

```python
# search(scoring)
def score_comparison(df_query, df_target, threshold = 18):
    dict_match = {}
    df_target_scored = df_target.copy()
    df_target_scored['score'] = 0

    for row in df_query.itertuples():
        for marker in standard_locus:
            df_target_scored['score'] = df_target_scored['score'] + df_target_scored[marker].apply(lambda x: 1 if x==getattr(row,marker) else 0)
        df_match = df_target_scored[df_target_scored['score']>=threshold]
        if len(df_match) > 0:
            dict_match[getattr(row,'Index')] = (df_match.index, df_match.score)
        df_target_scored['score'] = 0
    return dict_match

def search_DB(df_query, df_target, date_from, date_to, threshold = 18):
    df_query_filtered = filter_single_crosschecked(df_query)
    df_target_filtered = filter_date(df_target, date_from, date_to)
    df_target_filtered = filter_overlap(df_query_filtered, df_target_filtered)
    dict_match = score_comparison(df_query_filtered, df_target_filtered, threshold)
    # print result
    for idx_query, data_match in dict_match.items():
        idxs_match, scores_match = data_match
        print(f'query : \n{df_query_filtered.loc[idx_query]}')
        for i, idx_match in enumerate(idxs_match):
            print(f'Match : \n{df_target_filtered.loc[idx_match]}')
            print(f'Score : {scores_match.iloc[i]}')
    return dict_match
    
    
    
```
